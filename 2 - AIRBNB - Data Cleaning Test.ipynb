```python
import pandas as pd
cols = [
    'id',
    'property_type',
    'room_type',
    'amenities',
    'accommodates',
    'bathrooms',
    'bed_type',
    'cancellation_policy',
    'cleaning_fee',
    'city',
    'description',
    'first_review',
    'host_has_profile_pic',
    'host_identity_verified',
    'host_response_rate',
    'host_since',
    'instant_bookable',
    'last_review',
    'latitude',
    'longitude',
    'name',
    'neighbourhood',
    'number_of_reviews',
    'review_scores_rating',
    'thumbnail_url',
    'zipcode',
    'bedrooms',
    'beds']
datauji = pd.read_csv('D:/S2 IPB/Semester 2/Pembelajaran Mesin Statistika/UTS/test.csv', usecols=cols)
datauji.head()
datauji.shape[0]
```

```python
data = datauji.drop(columns=['description','first_review','host_response_rate','last_review','name','thumbnail_url'])
```

```python
data['review_scores_rating'] = data['review_scores_rating'].fillna(0)
meanbath = round(data['bathrooms'].mean())
data['bathrooms'] = data['bathrooms'].fillna(meanbath)
data['host_has_profile_pic'] = data['host_has_profile_pic'].fillna(0)
data['host_identity_verified'] = data['host_identity_verified'].fillna(0)
data['host_since'] = data['host_since'].fillna(0)
data['bedrooms'] = data['bedrooms'].fillna(0)
meanbed = round(data['beds'].mean())
data['beds'] = data['beds'].fillna(meanbed)
print(data.isnull().sum())
```

```python
data.host_has_profile_pic = data.host_has_profile_pic.replace({"t": 1, "f": 0})
data.host_identity_verified  = data.host_identity_verified.replace({"t": 1, "f": 0})
data.cleaning_fee  = data.cleaning_fee.replace({True: 1, False: 0})
data.instant_bookable   = data.instant_bookable.replace({"t": 1, "f": 0})
# fill zipcode yang kosong dengan kategori zipcode yang terbanyak 
data['zipcode'] = data['zipcode'].fillna(11211)

# fill neighbourhood yang ksong dengan kategori neighbourhood terbanyak
data['neighbourhood'] = data['neighbourhood'].fillna('Williamsburg')
print(data.isnull().sum())
```

```python
# Mengubah format date hosting since
from datetime import datetime
now = datetime.now()
now = now.strftime("%m/%d/%Y")

# Menbuat kolom baru waktu sekarang
data['days'] = now
data.info()
```

```python
# Mengubah format kolom host_since dan kolom days
data['host_since']= pd.to_datetime(data['host_since'])
data['days'] = pd.to_datetime(data['days'])
data.info()
```

```python
data['time'] = (data['days'] - data['host_since']).dt.days
data.head()
```

```python
data = data.drop('host_since', axis=1)
data = data.drop('days', axis=1)
print(data.isnull().sum())
```

```python
data.to_csv('uji.csv', index = False)
```
