```python
import pandas as pd
```

```python
datalatih = pd.read_csv('explorelatih.csv')
datalatihfix = datalatih.drop(columns=['zipcode','property_type','city','time','latitude','longitude','neighbourhood',
                                     'review_scores_rating'])
datalatihfix.head(20)
```

```python
# encode variabel kategori
room_dummies = pd.get_dummies(datalatihfix['room_type'])
bed_dummies = pd.get_dummies(datalatihfix['bed_type'])
cancel_dummies = pd.get_dummies(datalatihfix['cancellation_policy'])
```

```python
# replace the old columns with our new one-hot encoded ones
df = pd.concat((datalatihfix.drop(['room_type', 'bed_type','cancellation_policy'], axis=1), \
    room_dummies.astype(int), bed_dummies.astype(int), cancel_dummies.astype(int)), \
    axis=1)
```

```python
cols = list(df.columns.values)
idx = cols.index('price')
rearrange_cols = cols[:idx] + cols[idx+1:] + [cols[idx]]
df = df[rearrange_cols]
df.head()
```

```python
df.to_csv('datalatihfix.csv')
```
