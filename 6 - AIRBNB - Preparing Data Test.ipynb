```python
import pandas as pd
```

```python
data2 = pd.read_csv('C:/Users/user/Documents/AIRBNB PROJECT/uji.csv')
data2.head()
```

```python
# merapihkan data kolom zipcode
def fix_zip(series):
      return series.astype(str).str.split('-').str[0].str.zfill(5)
data2['zipcode'] = fix_zip(data2['zipcode'])
data2['zipcode'] = data2['zipcode'].str.replace('.', ' ')
data2["zipcode"] = data2["zipcode"].str.extract(r"(\d{5})")
```

```python
# EXPLORE DATA NUMERIK
## akan dilihat pada kolom berikut yang memiliki nilai nol, karena kolom dibawah ini tidak mungkin bernilai nol
count = (data2['accommodates'] == 0).sum()
count1 = (data2['bathrooms'] == 0).sum()
count2 = (data2['beds'] == 0).sum()
count3 = (data2['zipcode'] == 0).sum()

print('Count of zeros in Column accommodates : ', count)
print('Count of zeros in Column bathrooms : ', count1)
print('Count of zeros in Column beds : ', count2)
print('Count of zeros in Column zipcode : ', count3)
```

```python
# karena tidak ada data yang boleh dihilangkan, maka akan di replace
meanbath = data2["bathrooms"].mean()
data2['bathrooms'] = data2['bathrooms'].replace(['0'], 'meanbath')
data2.bathrooms = data2.bathrooms.round()
meanbed = round(data2["beds"].mean())
data2['beds'] = data2['beds'].replace(['0'], 'meanbed')
```

```python
# akan dihilangkan variabel bebas yang tidak digunakan
datafix = data2.drop(columns=['amenities','number_of_reviews','city','time','latitude','longitude','neighbourhood',
                                'review_scores_rating','zipcode','property_type'])
datafix.head(20)
```

```python
# encode variabel kategori
room_dummies = pd.get_dummies(datafix['room_type'])
bed_dummies = pd.get_dummies(datafix['bed_type'])
cancel_dummies = pd.get_dummies(datafix['cancellation_policy'])
```

```python
# replace the old columns with our new one-hot encoded ones
df = pd.concat((datafix.drop(['room_type', 'bed_type','cancellation_policy'], axis=1), \
    room_dummies.astype(int), bed_dummies.astype(int), cancel_dummies.astype(int)), \
    axis=1)
```
